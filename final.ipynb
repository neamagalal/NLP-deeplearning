# **reading data**

```python
import pandas as pd
data_train=pd.read_csv('pendigits-tra (1).csv')
data_test=pd.read_csv('pendigits-tes (1).csv')
x_train=data_train.iloc[: , :-1]
y_train=data_train.iloc[:, -1]
x_test=data_test.iloc[: , :-1]
y_test=data_test.iloc[:, -1]
```

# **decision tree**

```python
#Fitting Decision Tree classifier to the training set  
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
estimator = DecisionTreeClassifier(random_state=123)
estimator.fit(x_train, y_train)
predY = estimator.predict(x_test)
dtAccuracy = accuracy_score(y_test, predY)
print(dtAccuracy)
```

# **naive bayes**

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB( )
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
gbAccuracy = accuracy_score(y_test,y_pred)
print(gbAccuracy)
```

# **mlp**

```python
from sklearn.neural_network import MLPClassifier
mlp= MLPClassifier(random_state=123)
mlp.fit(x_train,y_train)
predict= mlp.predict(x_test)
mlpAccuracy = accuracy_score(y_test,predict)
print(mlpAccuracy)
```

# **svm**

```python
from sklearn import svm
clf = svm.SVC(random_state=123)
clf.fit(x_train,y_train)
svmpredict= clf.predict(x_test)
svmAccuracy = accuracy_score(y_test,svmpredict)
print(svmAccuracy)
```

# **gradient boost**

```python
from sklearn.ensemble import GradientBoostingClassifier
gbs = GradientBoostingClassifier(random_state=123).fit(x_train, y_train)
gbspredeict=gbs.predict(x_test)
gbsAccuracy = accuracy_score(y_test,gbspredeict)
print(gbsAccuracy)
```

```python
print("Decision tree accuracy:",dtAccuracy)
print("Naive Bayes accuracy:",gbAccuracy)
print("MLP accuracy:",mlpAccuracy)
print("SVM accuracy:",svmAccuracy)
print("Gradiant boost accuracy:",gbsAccuracy)
```

# **plot the base line**

```python
import pandas as pd
import matplotlib.pyplot as plot
df = pd.DataFrame({'classifiers':['decision tree', 'naive bayes','MLP','SVM','gradient boost'], 'accuracy_score':[dtAccuracy ,gbAccuracy , mlpAccuracy , svmAccuracy,gbsAccuracy]})
df.plot.line (x='classifiers', y='accuracy_score', rot=30,marker = "o")
df.plot.bar(x='classifiers', y='accuracy_score', rot=30)
plot.show(block=True);
```
